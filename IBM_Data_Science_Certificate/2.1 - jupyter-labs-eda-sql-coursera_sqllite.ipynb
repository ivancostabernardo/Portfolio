## Tasks

Now write and execute SQL queries to solve the assignment tasks.

**Note: If the column names are in mixed case enclose it in double quotes

### Task 1




##### Display the names of the unique launch sites  in the space mission


```python
%%sql
SELECT DISTINCT Launch_Site
FROM SPACEXTBL;
```


### Task 2


#####  Display 5 records where launch sites begin with the string 'CCA' 


```python
%%sql
SELECT *
FROM SPACEXTBL
WHERE Launch_Site LIKE 'CCA%'
LIMIT 5;
```

### Task 3




##### Display the total payload mass carried by boosters launched by NASA (CRS)


```python
%%sql
SELECT SUM(PAYLOAD_MASS__KG_) AS TOTAL_PAYLOAD_MASS
FROM SPACEXTBL
WHERE Customer = 'NASA (CRS)';
```

### Task 4




##### Display average payload mass carried by booster version F9 v1.1


```python
%%sql
SELECT AVG(PAYLOAD_MASS__KG_) AS AVG_PAYLOAD_MASS
FROM SPACEXTBL
WHERE Booster_Version = 'F9 v1.1';
```

### Task 5

##### Find the date when the first succesful landing outcome on the ground pad was acheived.


_Hint: Use min function_ 


```python
%%sql
SELECT MIN(Date) AS DATE_FIRST_SUCCESS_GROUND
FROM SPACEXTBL
WHERE Landing_Outcome = 'Success (ground pad)';
```

### Task 6

##### List the names of the boosters that have success in drone ship landing and have payload mass greater than 4000 but less than 6000

Firstly, we are going to find the exact name of the landing outcome we are interested in.

```python
%%sql
SELECT DISTINCT Landing_Outcome
FROM SPACEXTBL;
```

```python
%%sql
SELECT Booster_Version
FROM SPACEXTBL
WHERE Landing_Outcome = 'Success (drone ship)'
      AND `PAYLOAD_MASS__KG_` > 4000
      AND `PAYLOAD_MASS__KG_` < 6000;
```

### Task 7




##### List the total number of successful and failed mission outcomes


```python
%%sql
SELECT `Mission_Outcome`, COUNT(*) AS N_MISSIONS
FROM SPACEXTBL
GROUP BY `Mission_Outcome`;
```

```python
%%sql
SELECT CASE WHEN Mission_Outcome LIKE '%uccess%' THEN 'Success'
            WHEN Mission_Outcome LIKE '%ailure%' THEN 'Failure'
       END AS SUCCESS_FAILURE, COUNT(*) AS N_MISSIONS
FROM SPACEXTBL
GROUP BY SUCCESS_FAILURE;
```

### Task 8

##### List the   names of the booster versions that have carried the maximum payload mass. Use a subquery.


```python
%%sql
SELECT DISTINCT Booster_Version
FROM SPACEXTBL
WHERE PAYLOAD_MASS__KG_ = (SELECT MAX(PAYLOAD_MASS__KG_)
                           FROM SPACEXTBL);
```

### Task 9


##### List the month names of failed landing_outcomes in drone ship, booster versions and launch_site for the months in 2015.

**Note: SQLite does not support monthnames. So you need to use substr(Date, 4, 2) as month to get the months and substr(Date,7,4) = '2015' for year.**


```python
%%sql
SELECT Landing_Outcome,
       CASE WHEN SUBSTR(Date, 4, 2) = '01' THEN 'January'
            WHEN SUBSTR(Date, 4, 2) = '02' THEN 'February'
            WHEN SUBSTR(Date, 4, 2) = '03' THEN 'March'
            WHEN SUBSTR(Date, 4, 2) = '04' THEN 'April'
            WHEN SUBSTR(Date, 4, 2) = '05' THEN 'May'
            WHEN SUBSTR(Date, 4, 2) = '06' THEN 'June'
            WHEN SUBSTR(Date, 4, 2) = '07' THEN 'July'
            WHEN SUBSTR(Date, 4, 2) = '08' THEN 'August'
            WHEN SUBSTR(Date, 4, 2) = '09' THEN 'September'
            WHEN SUBSTR(Date, 4, 2) = '10' THEN 'October'
            WHEN SUBSTR(Date, 4, 2) = '11' THEN 'November'
            WHEN SUBSTR(Date, 4, 2) = '12' THEN 'December'
       END AS MONTH_NAME, 
       Booster_Version, 
       Launch_Site
FROM SPACEXTBL
WHERE Landing_Outcome = 'Failure (drone ship)'
      AND SUBSTR(DATE, 7, 4) = '2015';
```

### Task 10

##### Rank the  count of Landing_Outcomes between the dates 2010-06-04 and 2017-03-20 in descending order.


```python
%%sql
SELECT Landing_Outcome, COUNT(*) AS N
FROM SPACEXTBL
WHERE SUBSTR(Date, 7, 4)||SUBSTR(Date, 4, 2)||SUBSTR(Date, 1, 2) BETWEEN '20100604' AND '20170320'
GROUP BY Landing_Outcome
ORDER BY N DESC;
```

```python
%%sql
SELECT STRFTIME('%d/%m/%Y')
FROM SPACEXTBL
WHERE STRFTIME('%d/%m/%Y', Date) = STRFTIME('%Y-%m-%d', '2010-04-06');
```
